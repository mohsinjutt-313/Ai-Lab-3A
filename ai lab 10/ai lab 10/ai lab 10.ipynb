### Filling Null values

```python
# compute modes for object columns (useful for filling)
modes = {col: df[col].mode().iat[0] if not df[col].mode().empty else None for col in df.select_dtypes(include='object').columns}
print(modes)
```

```python
# Fill missing values: numeric -> mean, object -> mode
for col in df.columns:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            mode = df[col].mode()
            fill = mode.iat[0] if not mode.empty else 'Unknown'
            df[col].fillna(fill, inplace=True)

print('Missing values remaining:', df.isnull().sum().sum())
```

```python
# show a quick numeric summary after filling
print(df.select_dtypes(include=['number']).describe())
```

### TYpe conversion to int

```python
import numpy as np

# Safely convert object columns that contain numeric values to numeric dtype.
# This handles commas and coerces non-convertible values to NaN.
for col in df.select_dtypes(include=['object']).columns:
    cleaned = df[col].astype(str).str.replace(',', '').str.strip()
    coerced = pd.to_numeric(cleaned, errors='coerce')
    # convert column if conversion produced mostly numbers (no loss)
    if coerced.notna().sum() >= (0.9 * len(coerced)):
        df[col] = coerced

print('Converted possible numeric-like object columns.')
print(df.dtypes)
```

### Converting objects columns into int columns

```python
# Final pass: convert remaining object columns that are purely numeric to integers
for col in df.select_dtypes(include=['object']).columns:
    if df[col].str.isnumeric().all():
        df[col] = df[col].astype(int)

print('Dtypes after final conversion:')
print(df.dtypes)
```
