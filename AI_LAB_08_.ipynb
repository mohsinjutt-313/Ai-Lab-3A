  Min-Max Algorithm

```python
import math

def minimax(depth, index, isMax, scores, targetDepth):
    # base case: reached the leaf level
    if depth == targetDepth:
        return scores[index]

    if isMax:
        a = minimax(depth + 1, index * 2, False, scores, targetDepth)
        b = minimax(depth + 1, index * 2 + 1, False, scores, targetDepth)
        return max(a, b)
    else:
        a = minimax(depth + 1, index * 2, True, scores, targetDepth)
        b = minimax(depth + 1, index * 2 + 1, True, scores, targetDepth)
        return min(a, b)

scores = [3, 5, 2, 9, 3, 5, 2, 9]
targetDepth = int(math.log(len(scores), 2))      # this must be 3 for 8 elements

result = minimax(0, 0, True, scores, targetDepth)
print("Optimal Value:", result)
```
