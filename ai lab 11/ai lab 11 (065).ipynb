### Train test splitting

```python
import pandas as pd
df = pd.read_csv('crop_yield.csv')

# Data Preprocessing
df = df.drop('Crop_Year', axis=1)  # Drop unnecessary column

# Remove trailing whitespace from object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

df.columns
```

```python
# Display shapes before train-test split
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False) 
```

```python
print(f"Training set shape: {train_X.shape}, {train_y.shape}")
print(f"Test set shape: {test_X.shape}, {test_y.shape}")
```

```python
print(train_X.head(2))
```

### Model training

```python
# Check data types before encoding
print(train_X.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}

for col in train_X.columns:
    if train_X[col].dtype == 'object':
        le = LabelEncoder()
        train_X[col] = le.fit_transform(train_X[col])
        test_X[col] = le.transform(test_X[col])
        encoders[col] = le

print("Encoding completed for categorical variables")
```

```python
# Check encoded data
print(train_X.head())
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(train_X, train_y)

print(model_rf)
```

```python
# Encoding already completed - verify encoded features
print("Categorical variables encoded successfully")
print(train_X.dtypes)
```

```python
print(f"Target variable dtype: {train_y.dtype}")
print(f"Target variable is numeric, no encoding needed")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions
y_pred_train = model_rf.predict(train_X)
y_pred_test = model_rf.predict(test_X)

# Evaluate model
train_r2 = r2_score(train_y, y_pred_train)
test_r2 = r2_score(test_y, y_pred_test)

print(f"Train R² Score: {train_r2:.4f}")
print(f"Test R² Score: {test_r2:.4f}")
```

```python
# Model already trained - check model parameters
print("RandomForest model trained successfully")
print(f"Number of trees: {model_rf.n_estimators}")
print(f"Feature importance:\n{pd.Series(model_rf.feature_importances_, index=train_X.columns).sort_values(ascending=False)}")
```
